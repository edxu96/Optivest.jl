```python
import pandas as pd
from io import StringIO

flows_ptes = pd.DataFrame({
    'source': ['Field3', 'PTES', 'PTES', 'Tank', 'PTES', 'Energy-Change'], 
    'target': ['PTES', 'Heat-Pump', 'Tank', 'PTES', 'Loss', 'PTES'], 
    'value': [10141, 7733, 7733, 10141, 4615, 212],
    'type': [1, 2, 3, 4, 0, 0]
})

flows_ptes
```

```python
from floweaver import *

# Set the default size to fit the documentation better.
size = dict(width=570, height=300)

nodes = {
    'input': ProcessGroup(['Field3', 'Tank', 'Energy-Change']),
    'PTES': ProcessGroup(['PTES']),
    'output': ProcessGroup(['Heat-Pump', 'Tank', 'Loss']),
}

nodes['input'].partition = Partition.Simple('process', [
    'Field3', 'Tank', 'Energy-Change'
])

nodes['output'].partition = Partition.Simple('process', [
    'Heat-Pump', 'Tank', 'Loss'
])

bundles = [
    Bundle('input', 'PTES'),
    Bundle('PTES', 'output'),
]

ordering = [
    ['input'],
    ['PTES'],
    ['output'],
]

sdd = SankeyDefinition(nodes, bundles, ordering)

weave(sdd, flows_ptes).to_widget(**size)
```
